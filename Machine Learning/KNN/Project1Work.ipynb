```python
import pandas
import numpy

bike = pandas.read_csv('bikeshare_projectA.csv')
# needs to be regression
def regression_kNN(new_example,train_data,k):
    #getting a cope of the training set just so we don't
    #mess up the original
    train_data_copy = train_data.copy()
    train_data_copy['distance_to_new'] = numpy.sqrt(
        (new_example['season']-train_data_copy['season'])**2
        +(new_example['mnth']-train_data_copy['mnth'])**2
        +(new_example['weekday']-train_data_copy['weekday'])**2
        +(new_example['weathersit']-train_data_copy['weathersit'])**2
        +(new_example['tempc']-train_data_copy['tempc'])**2
        +(new_example['windspeed']-train_data_copy['windspeed'])**2)
    sorted_data = train_data_copy.sort_values(['distance_to_new'])
    #mode to get most common thing in the first k examples in the sorted dataframe
    #iloc to get the actual string, mode will return the string inside of a pandas Series
    prediction = sorted_data.iloc[0:k]['cnt'].mean()
    # regression = 
    # return regression
    return prediction 
    

def regression_all_kNN(test_data,train_data,k):
    #apply the classify_kNN function to each item in the test data with the train
    #data and k passed as the other two arguments. The result will be a series of the individual results.
    return test_data.apply(regression_kNN,args=(train_data,k),axis=1)

def MAE(actual, predicted):
    mae = abs(actual - predicted)
    return mae
    ##def accuracy(actual,predicted):
    #get the series comparing the two series
```

```python
def normbike(bike):
    
    bike2= bike.copy()
    
    #calculate mean and std for season
    s_mean = bike2['season'].mean()
    s_std = bike2['season'].std()
    #replace column with each entry’s z-score
    bike2['season'] = (bike2['season']-s_mean)/s_std

    #calculate mean and std for mnth
    m_mean = bike2['mnth'].mean()
    m_std = bike2['mnth'].std()
    #replace column with each entry’s z-score
    bike2['mnth'] = (bike2['mnth']-m_mean)/m_std

    #calculate mean and std for weekday
    w_mean = bike2['weekday'].mean()
    w_std = bike2['weekday'].std()
    #replace column with each entry’s z-score
    bike2['weekday'] = (bike2['weekday']-w_mean)/w_std

    #calculate mean and std for weathersit
    wes_mean = bike2['weathersit'].mean()
    wes_std = bike2['weathersit'].std()
    #replace column with each entry’s z-score
    bike2['weathersit'] = (bike2['weathersit']-wes_mean)/wes_std

    #calculate mean and std for tempc
    t_mean = bike2['tempc'].mean()
    t_std = bike2['tempc'].std()
    #replace column with each entry’s z-score
    bike2['tempc'] = (bike2['tempc']-t_mean)/t_std

    #calculate mean and std for windspeed
    wis_mean = bike2['windspeed'].mean()
    wis_std = bike2['windspeed'].std()
    #replace column with each entry’s z-score
    bike2['windspeed'] = (bike2['windspeed']-wis_mean)/wis_std
    
    return bike2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#reload the data
bike = pandas.read_csv('bikeshare_projectA.csv')
#normalized data

shuffled_data = bike.sample(frac=1)
shuffled_data_norm = normbike(shuffled_data)

#original
test_data = shuffled_data.iloc[0:100]
train_data = shuffled_data.iloc[100:]

#normalized
test_data_norm = shuffled_data_norm.iloc[0:100]
train_data_norm = shuffled_data_norm.iloc[100:]

k_vals = [1,3,5,9,15,50,100,200,500,700]
kNN_Maes = []
kNN_Maes_norm = []


for k in k_vals:

    predictions = regression_all_kNN(test_data,train_data,k)
    predictions_norm = regression_all_kNN(test_data_norm,train_data_norm,k)
    
    current_MAE = sum(MAE(test_data['cnt'],predictions))/100
    current_MAE_norm = sum(MAE(test_data_norm['cnt'],predictions_norm))/100
    
    kNN_Maes.append(current_MAE)
    kNN_Maes_norm.append(current_MAE_norm)


plt.suptitle('Bike Data k-NN Experiment',fontsize=18)
plt.xlabel('k')
plt.ylabel('Mean Absolute Error')
plt.plot(k_vals,kNN_Maes,'ro-',label='k-NN')
plt.plot(k_vals,kNN_Maes_norm,'bs-',label='k-NN Normalized')
plt.legend(loc='lower left', shadow=True)
plt.axis([0,750,0,1650])

plt.show()
```
